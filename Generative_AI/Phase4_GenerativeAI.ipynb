# Phase 4 — Generative AI Integration (Simulation Only)

This phase simulates the behavior of a Generative AI component using the clustering results from Phase 3.

Objectives:
- Use clustering results
- Two prompt templates
- Generate recommendations
- Compare and pick the best


```
import pandas as pd

cluster_labels = {0: 'High Performance', 1: 'Moderate Performance', 2: 'Low Performance'}

students = [
    {'student_id': 'S1', 'cluster_id': 2, 'math_score': 45, 'reading_score': 50, 'writing_score': 48},
    {'student_id': 'S2', 'cluster_id': 1, 'math_score': 70, 'reading_score': 72, 'writing_score': 68},
    {'student_id': 'S3', 'cluster_id': 0, 'math_score': 92, 'reading_score': 88, 'writing_score': 90}
]

students_df = pd.DataFrame(students)
students_df['cluster_label'] = students_df['cluster_id'].map(cluster_labels)
students_df
```

## Prompt Templates

```
template_a = 'General study recommendations.'
template_b = 'Subject-based weekly improvement plan.'

template_a, template_b
```

```
def describe_cluster(cluster_id):
    if cluster_id == 0:
        return 'High performing. Can handle advanced tasks.'
    elif cluster_id == 1:
        return 'Moderate performance. Needs improvement.'
    return 'Low performance. Needs structured support.'

def advice_template_a(student):
    return f"A | {student['student_id']}: {describe_cluster(student['cluster_id'])} Study daily."

def advice_template_b(student):
    return f"B | {student['student_id']}: {describe_cluster(student['cluster_id'])} Weekly Plan: Practice subjects."
```

```
results = []
for student in students:
    results.append(advice_template_a(student))
    results.append(advice_template_b(student))
results
```

## Comparison
- Template A → Simple
- Template B → Actionable
### Winner: Template B
